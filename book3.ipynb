```python
import xlrd
from collections import Counter
```

```python
from xlrd import open_workbook, cellname
book1 = open_workbook('exits.xlsx')
book2 = open_workbook('Book2.xlsx')
sheet1 = book1.sheet_by_index(0)
sheet2 = book2.sheet_by_index(0)
print (sheet1.name)
print (sheet1.nrows)
print (sheet1.ncols)
print (sheet2.name)
print (sheet2.nrows)
print (sheet2.ncols)
```

```python
ramp_w90 = []
ramp_e90 = []
ramp_w94 = []
ramp_e94 = []
ramp_n294 = []
ramp_s294 = []
ramp_w88 = []
ramp_e88 = []
ramp_n355 = []
ramp_s355 = []
ramp_w390 = []
ramp_e390 = []
for row in range(sheet1.nrows):
    value_w90 = []
    value_w94 = []
    value_n294 = []
    value_w88 = []
    value_n355 = []
    value_w390 = []
    value_e90 = []
    value_e94 = []
    value_s294 = []
    value_e88 = []
    value_s355 = []
    value_e390 = []   
    if sheet1.cell(row,0).value== 1:
        for col in range(sheet1.ncols):
            value_w90.append(sheet1.cell(row,col).value)           
        ramp_w90.append(value_w90)
    lif sheet1.cell(row,1).value== 2:
        for col in range(sheet1.ncols):
            value_e90.append(sheet1.cell(row,col).value)    
        ramp_e90.append(value_e90)
    elif sheet1.cell(row,1).value== 3:
        for col in range(sheet1.ncols):
            value_w94.append(sheet1.cell(row,col).value)    
        ramp_w94.append(value_w94)
    elif sheet1.cell(row,1).value== 4:
        for col in range(sheet1.ncols):
            value_e94.append(sheet1.cell(row,col).value)    
        ramp_e94.append(value_e94)
    elif sheet1.cell(row,1).value== 5:
        for col in range(sheet1.ncols):
            value_n294.append(sheet1.cell(row,col).value)    
        ramp_n294.append(value_n294)
    elif sheet1.cell(row,1).value== 6:
        for col in range(sheet1.ncols):
            value_s294.append(sheet1.cell(row,col).value)    
        ramp_s294.append(value_s294)
    elif sheet1.cell(row,1).value== 9:
        for col in range(sheet1.ncols):
            value_w88.append(sheet1.cell(row,col).value)    
        ramp_w88.append(value_w88)
    elif sheet1.cell(row,1).value== 10:
        for col in range(sheet1.ncols):
            value_e88.append(sheet1.cell(row,col).value)    
        ramp_e88.append(value_e88)
    elif sheet1.cell(row,1).value== 11:
        for col in range(sheet1.ncols):
            value_n355.append(sheet1.cell(row,col).value)    
        ramp_n355.append(value_n355)
    elif sheet1.cell(row,1).value== 12:
        for col in range(sheet1.ncols):
            value_s355.append(sheet1.cell(row,col).value)    
        ramp_s355.append(value_s355)
    elif sheet1.cell(row,1).value== 19:
        for col in range(sheet1.ncols):
            value_w390.append(sheet1.cell(row,col).value)    
        ramp_w390.append(value_w390)
    elif sheet1.cell(row,1).value== 20:
        for col in range(sheet1.ncols):
            value_e390.append(sheet1.cell(row,col).value)    
        ramp_e390.append(value_e390)
print(ramp_e90)
```

```python
site_e90 = []
site_e94 = []
site_n294 = []
site_e88 = []
site_n355 = []
site_e390 = []
site_w90 = []
site_w94 = []
site_s294 = []
site_w88 = []
site_s355 = []
site_w390 = []
for row in range(sheet2.nrows):
    value_w90 = []
    value_w94 = []
    value_n294 = []
    value_w88 = []
    value_n355 = []
    value_w390 = []
    value_e90 = []
    value_e94 = []
    value_s294 = []
    value_e88 = []
    value_s355 = []
    value_e390 = [] 

    if sheet2.cell(row,1).value=='I-90'and sheet2.cell(row,4).value=='WB':
        for col in range(sheet2.ncols):
            value_w90.append(sheet2.cell(row,col).value) 
        site_w90.append(value_w90)   
    elif sheet2.cell(row,1).value=='I-90'and sheet2.cell(row,4).value=='EB':
        for col in range(sheet2.ncols):
            value_e90.append(sheet2.cell(row,col).value)    
        site_e90.append(value_e90)
    elif sheet2.cell(row,1).value=='I-94'and sheet2.cell(row,4).value=='WB':
        for col in range(sheet2.ncols):
            value_w94.append(sheet2.cell(row,col).value)    
        site_w94.append(value_w94)
    elif sheet2.cell(row,1).value=='I-94'and sheet2.cell(row,4).value=='EB':
        for col in range(sheet2.ncols):
            value_e94.append(sheet2.cell(row,col).value)    
        site_e94.append(value_e94)
    elif sheet2.cell(row,1).value=='I-294'and sheet2.cell(row,4).value=='NB':
        for col in range(sheet2.ncols):
            value_n294.append(sheet2.cell(row,col).value)    
        site_n294.append(value_n294)
    elif sheet2.cell(row,1).value=='I-294'and sheet2.cell(row,4).value=='SB':
        for col in range(sheet2.ncols):
            value_s294.append(sheet2.cell(row,col).value)    
        site_s294.append(value_s294)
    elif sheet2.cell(row,1).value=='I-88'and sheet2.cell(row,4).value=='EB':
        for col in range(sheet2.ncols):
            value_e94.append(sheet2.cell(row,col).value)    
        site_e90.append(value_e90)
    elif sheet2.cell(row,1).value=='I-88'and sheet2.cell(row,4).value=='WB':
        for col in range(sheet2.ncols):
            value_w88.append(sheet2.cell(row,col).value)    
        site_w88.append(value_w88)
    elif sheet2.cell(row,1).value=='I-355'and sheet2.cell(row,4).value=='SB':
        for col in range(sheet2.ncols):
            value_s355.append(sheet2.cell(row,col).value)    
        site_s355.append(value_s355)
    elif sheet2.cell(row,1).value=='I-355'and sheet2.cell(row,4).value=='NB':
        for col in range(sheet2.ncols):
            value_n355.append(sheet2.cell(row,col).value)    
        site_n355.append(value_n355)
    elif sheet2.cell(row,1).value=='IL-390'and sheet2.cell(row,4).value=='WB':
        for col in range(sheet2.ncols):
            value_w390.append(sheet2.cell(row,col).value)    
        site_w390.append(value_s294)
    elif sheet2.cell(row,1).value=='IL-390'and sheet2.cell(row,4).value=='EB':
        for col in range(sheet2.ncols):
            value_e390.append(sheet2.cell(row,col).value)    
        site_e390.append(value_e390)
```

```python
order_w90 =[]
hotspot_w90 = []
for i in range(len(site_w90)):
    res = [sub[2] for sub in ramp_w90] 
    res.append(site_w90[i][5])
    temp = sorted(res)
    num = temp.index(site_w90[i][5])
    hotspot_w90.append(ramp_w90[num-1][4])
    hotspots_w90 = Counter(hotspot_w90)
sss = [ss[8] for ss in site_w90]
#print(sss)
print(hotspots_w90)

```
